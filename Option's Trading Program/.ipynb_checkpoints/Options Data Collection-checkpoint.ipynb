```python
stock_data = response.json()

stock_df = pd.DataFrame()

for s,i in enumerate(stock_data.get('optionChain', {}).get('result', {})[0].get('options', [])[0].get('calls', [])):
    
    stock_df = stock_df.append(pd.DataFrame.from_dict(i, orient = 'index').transpose())

stock_df = stock_df.reset_index(drop = True).fillna(0)
```
